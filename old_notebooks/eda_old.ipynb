```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

### Emissions Data (greenhouse gases)

By country (including world), annual readings. 1850-2022.

```python
emissions = pd.read_csv('co2-yearly.csv')
print(f'Rows: {emissions.shape[0]}')
emissions
```

```python
# Select global data
# Columns of interest are 'co2', 'population', 'year', and maybe 'gdp'
emissions_world = emissions[emissions['country'] == 'World']
emissions_world
```

```python
emissions_world.plot(kind='line', x='year', y='co2')
```

### Electricity Generation data

By country (including world). Annual readings. 2000-2021.

```python
electricity = pd.read_csv('electricity-yearly.csv')
print(f'Rows: {electricity.shape[0]}')
electricity
```

```python
electricity_world = electricity[electricity['Entity'] == 'World']
electricity_world
```

```python
electricity_world.plot(kind='line', x='Year', y='Electricity generation - TWh')
```

### Change in Global Mean Temperature (Celsius)

World data only. Annual readings. 1880-2023.

```python
global_temp = None

with open('global-temp-change-yearly.txt') as f:
    lines = f.readlines()[5:]
    lines = [' '.join(line.split()) for line in lines]

    data = np.array([line.split(' ') for line in lines], dtype=np.float32)
    
    global_temp = pd.DataFrame(columns=['year', 'temp_change', 'temp_change_smoothed'], data=data)

global_temp
```

```python
global_temp.plot(kind='line', x='year', y='temp_change')
```
